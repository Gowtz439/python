# POPULARITY VS YEAR

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='year', y='popularity')
plt.title('Popularity of Tracks Over the Years')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.show()
```

# DANCEABILITY VS TEMPO

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='tempo', y='danceability')
plt.title('Danceability vs. Tempo')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Danceability')
plt.show()
```

# LOUDNESS VS ENERGY

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='loudness', y='energy')
plt.title('Loudness vs. Energy')
plt.xlabel('Loudness')
plt.ylabel('Energy')
plt.show()
```


### 1. Popularity of Tracks Over the Years
This line plot shows how the popularity of tracks has changed over the years. It helps in understanding trends in music popularity over time.

**Analysis**:
- The graph indicates fluctuations in the popularity of tracks over different years.
- Peaks in popularity might correspond to the release of hit songs or albums.
- Understanding these trends can help in predicting future popular tracks and identifying key factors that contribute to a track's success.

### 2. Danceability vs. Tempo
This scatter plot illustrates the relationship between danceability and tempo of the tracks. It can help identify how tempo influences the danceability of a song.

**Analysis**:
- There is a noticeable correlation between tempo and danceability.
- Tracks with higher tempos tend to have higher danceability scores, making them more suitable for dancing.
- This information can be useful for music producers and DJs when creating or selecting tracks for dance events.

### 3. Loudness vs. Energy
This scatter plot shows the relationship between loudness and energy of the tracks. It helps in understanding how these two attributes correlate with each other.

**Analysis**:
- The graph shows a positive correlation between loudness and energy.
- Tracks with higher loudness levels tend to have higher energy, making them more dynamic and engaging.
- This insight can be valuable for audio engineers and music producers aiming to create high-energy tracks.
