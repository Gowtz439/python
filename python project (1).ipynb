```python
import pandas as pd

df=pd.read_csv("spotify_tracks.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Identify duplicated rows
duplicated_rows = df[df.duplicated()]

# Print the columns of duplicated rows
print(duplicated_rows)
```

```python
df.drop('track_url', axis=1, inplace=True)
```

```python
df.drop('time_signature', axis=1, inplace=True)
```

```python
df.drop('track_id', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop('artwork_url', axis=1, inplace=True)
```

```python
df.head()
```

```python
df=df[df['year'] > 2005]
df=df[df['energy'] > 0]
```

# language vs count of songs

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='language')
plt.title('language vs count of songs')
plt.xlabel('languages')
plt.ylabel('songs count')
plt.show()
```

# language vs. popularity(%)

```python

plt.figure(figsize=(8, 4))
sns.barplot(x='language', y='popularity', data=df)
plt.title('language vs. popularity(%)')
plt.xlabel('language')
plt.ylabel('popularity(%)')
plt.show()
```

# year vs number of songs

```python

plt.figure(figsize=(15,4))
sns.countplot(data=df, x='year')
plt.title('year vs number of songs')
plt.xlabel('year')
plt.ylabel('songs count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='danceability', bins=30, kde=True)
plt.title('Distribution of danceability')
plt.xlabel('danceability(score)')
plt.ylabel('count')
plt.show()

```

Visualizations and Analysis
1. Language vs. Count of Songs
This visualization shows the distribution of songs across different languages. It helps identify which languages are most represented in your dataset.

2. Language vs. Popularity (%)
This bar plot compares the average popularity of songs across different languages. It provides insights into which languages have more popular songs.

3. Year vs. Number of Songs
This count plot shows the number of songs released each year. It helps identify trends in music production over time.

4. Danceability distribution
The danceability score ranges from 0.0 to 1.0, where higher values indicate that the track is more danceable
